# **Árvore B** > *Especialista em Fundamentos: Eduardo Artur*

A Árvore B é uma estrutura de dados usada para organizar grandes quantidades de informações de forma rápida e eficiente.

Ela funciona como uma árvore onde cada bloco (nó) pode guardar vários valores ao mesmo tempo.

Exemplo:
                      

                     [30|60]  
                   /    |    \               
             [10|20] [40|50] [70|80]

Isso ajuda a diminuir a altura da árvore e acelerar buscas.

Propriedades:
- Todos os nós folha ficam no mesmo nível;
- Cada nó possui um limite mínimo e máximo de valores;
- A árvore sempre permanece balanceada;
- As operações de busca, inserção e remoção são rápidas, com complexidade O(log n).

Regras Básicas:
- Os valores dentro do nó ficam ordenados;
- Máximo de M filhos por nó;
- Máximo de M − 1 chaves por nó;
- Quando o limite é ultrapassado ocorre um Split.

# **Árvore B+**

Definição da Estrutura

A Árvore B+ é uma versão melhorada da Árvore B.
Árvore B+

                  [30|60]
               /     |     \
          [10,20]↔[30,50]↔[60,80]

Nela:

- os dados ficam apenas nos nós folha;
- os nós internos servem apenas para indicar caminhos;
- as folhas ficam ligadas entre si.

Isso torna buscas sequenciais muito mais rápidas.

Propriedades:
- Todas as folhas ficam no mesmo nível;
- A árvore permanece balanceada;
- As operações possuem complexidade: O(logn)
- Os nós folha armazenam todos os registros.

Regras Básicas
- Os dados reais ficam somente nas folhas;
- Os nós internos funcionam como índices;
- As folhas são conectadas em sequência;
- Quando um nó enche, ocorre divisão (split);
- A estrutura permanece organizada automaticamente.


# **Comparando: Árvore B vs Árvore B+**

| Característica | Árvore B | Árvore B+ |
|---|---|---|
| Onde ficam os dados? | Em **todos** os nós (internos e folhas) | Apenas nos **nós folha** |
| Nós internos | Guardam chaves **e dados** | Funcionam apenas como **índice** (placa de trânsito) |
| Busca por um valor único | Pode ser mais rápida se o dado estiver na raiz | Sempre desce até a folha |
| Busca em intervalo (ex: de 10 a 50) | Difícil — precisa subir e descer na árvore | **Muito rápida** — folhas conectadas em sequência |
| Onde é mais usada? | Sistemas de arquivos (macOS, Windows) | **Bancos de dados** (MySQL, PostgreSQL) |

> 💡 **Resumo simples:** A B+ é melhor para buscas em grande volume de dados porque os dados ficam todos nas folhas, conectadas como uma fila. Você desce uma vez e caminha reto.

# **Visualizando o Split (Divisão) na Árvore B** > *Analista de Operações: Joseildo*

Vamos simular passo a passo o que acontece quando uma Árvore B precisa se reorganizar.

Neste exemplo, cada nó comporta no máximo **2 valores**.

## Passo 1 — Inserindo 10 e 20

A árvore começa vazia. Inserimos o 10 e depois o 20.

Como ainda existe espaço disponível no nó, nenhuma reorganização é necessária.

```text id="5st4v1"
[ 10 | 20 ]
```



## Passo 2 — Problema! Inserindo 30

O nó já está cheio com os valores 10 e 20.

Ao inserir o valor 30, o limite é ultrapassado e ocorre um **Split (Divisão)**.

```text id="mlrx1f"
[ 10 | 20 | 30 ]   ← CHEIO! Vai ser dividido.
```



## Passo 3 — Resultado do Split

O valor central (20) sobe para o nó pai.

Os valores menores ficam à esquerda e os maiores ficam à direita.

```text id="fjlwmu"
        [ 20 ]
       /      \
    [ 10 ]  [ 30 ]
```

🔎 O valor **20** tornou-se o nó guia. Toda busca menor que 20 segue para a esquerda, e toda busca maior que 20 segue para a direita.



## Passo 4 — Inserindo 40

O valor 40 é maior que 20, então é inserido no nó da direita.

```text id="mr7xdr"
        [ 20 ]
       /      \
    [ 10 ] [ 30 | 40 ]
```



## Passo 5 — Novo Problema! Inserindo 50

O nó da direita já está cheio. Ao inserir 50 ocorre um novo Split.

```text id="95hwjw"
        [ 20 ]
       /      \
    [ 10 ] [ 30 | 40 | 50 ]
                     ↑
                  CHEIO!
```



## Passo 6 — Resultado do Segundo Split

O valor central (40) sobe para o nó pai.

```text id="p8v8gl"
          [ 20 | 40 ]
         /     |     \
      [10]   [30]   [50]
```

🔎 Agora a árvore possui dois valores guia (20 e 40), permitindo localizar rapidamente em qual ramo a busca deve continuar.

### Resumo do Split

Sempre que um nó ultrapassa sua capacidade:

1. O valor do meio sobe para o nó pai.
2. Os valores menores ficam no filho esquerdo.
3. Os valores maiores ficam no filho direito.
4. A árvore permanece balanceada.

Exemplo:

```text id="lt8xg9"
Antes:

[ 10 | 20 | 30 ]

Depois:

       [ 20 ]
      /      \
   [10]    [30]
```

```
[ 10 | 20 ]
```



### Passo 2 — Problema! Inserindo 30

O bloco já está cheio com `[10 | 20]` e queremos inserir o 30.
Não tem mais espaço! É aqui que acontece o **Split (divisão)**.

```
[ 10 | 20 | 30 ]  ← CHEIO! Vai partir ao meio.
```

O valor do meio **sobe** para virar o nó guia (nó interno). Os dados continuam nas folhas, conectadas entre si:

```
         [ 20 ]           ← Nó Interno (só serve de placa de trânsito)
        /       \\
  [ 10 ]  --->  [ 20 | 30 ]  ← Nós Folha (onde os dados realmente ficam)
```

> 🔎 Repare que o **20** subiu para guiar a busca. Se você procurar um número menor que 20, vai para a esquerda. Se for maior ou igual, vai para a direita. E as folhas continuam conectadas como uma fila!

## **Árvore B**

### Melhor Caso

A árvore permanece balanceada e possui vários valores por nó.

A altura da árvore é aproximadamente:

```text
h ≈ logₘ(n)
```

onde m representa a ordem da árvore.

Complexidade:

* Busca: O(log n)
* Inserção: O(log n)
* Remoção: O(log n)


### Pior Caso

Mesmo com vários splits e reorganizações, a árvore continua balanceada.

Complexidade:

* Busca: O(log n)
* Inserção: O(log n)
* Remoção: O(log n)

```text id="q4n7ak"
        [30 | 60]
      /     |     \
 [10|20] [40|50] [70|80]
```

## **Árvore B+**

### Melhor Caso

Os dados ficam organizados nas folhas e a árvore permanece balanceada.

Complexidade:

* Busca: O(log n)
* Inserção: O(log n)
* Remoção: O(log n)

### Pior Caso

Mesmo ocorrendo divisões (splits) e redistribuições de nós, a estrutura permanece balanceada.

Complexidade:

* Busca: O(log n)
* Inserção: O(log n)
* Remoção: O(log n)

# **Onde Isso É Usado no Mundo Real?**

Sistemas gigantes como **MySQL** (engine InnoDB) e **PostgreSQL** usam a Árvore B+ por baixo dos panos para criar os **índices** das tabelas.

### O Exemplo Prático

Imagine um banco de dados do **Nubank** com 50 milhões de clientes.

Se você faz a consulta:
```sql
SELECT * FROM clientes WHERE cpf = '123.456.789-00'
```

O banco **não lê 50 milhões de linhas** uma por uma. Ele usa a Árvore B+: olha para a raiz, desce pelos nós internos e, em **3 ou 4 pulos**, cai direto no bloco do HD onde está o CPF buscado.

### A Busca em Intervalo no SQL

```sql
SELECT * FROM clientes WHERE idade BETWEEN 20 AND 30
```

- **Árvore B:** o banco teria que subir e descer pelos galhos para achar cada idade.
- **Árvore B+:** desce uma vez para achar o `20` e depois **anda reto pelas folhas conectadas** até chegar no `30`. Muito mais rápido!

---

| Onde é Usada | Árvore | Por quê? |
|---|---|---|
| MySQL (InnoDB), PostgreSQL | **B+** | Índices de tabelas e buscas em intervalo |
| Sistema de arquivos NTFS (Windows) | **B** | Organização de metadados de arquivos |
| Sistema de arquivos HFS+ (macOS) | **B+** | Diretórios e índices de arquivos |
| MongoDB (WiredTiger) | **B+** | Índices de coleções |

Agora vamos colocar a mão na massa! O código abaixo implementa uma **Árvore B completa e funcional** com as operações de:

* Inserção (com *Split* automático quando um nó enche);
* Busca por um valor específico;
* Percurso ordenado dos elementos;
* Balanceamento automático da estrutura.

A Árvore B é amplamente utilizada em bancos de dados e sistemas de arquivos por manter baixo o número de acessos necessários para localizar informações, mesmo quando a quantidade de dados é muito grande.

## **Demonstração Guiada do Código: Árvore B** > *Instrutor de Prática: Reginaldo*

````python
# ==========================================
# CLASSE QUE REPRESENTA UM NÓ DA ÁRVORE B
# ==========================================

class NoArvoreB:
    def __init__(self, folha=False):
        self.folha = folha          # Indica se o nó é folha
        self.chaves = []            # Valores armazenados no nó
        self.filhos = []            # Ponteiros para os filhos


# ==========================================
# CLASSE DA ÁRVORE B
# ==========================================

class ArvoreB:
    def __init__(self, grau_minimo=2):
        """
        grau_minimo = 2

        Isso significa que a árvore será de ordem 4,
        podendo ter até:
        - 4 filhos por nó
        - 3 chaves por nó
        """
        self.raiz = NoArvoreB(True)
        self.grau_minimo = grau_minimo


    # ==========================================
    # BUSCA UM VALOR NA ÁRVORE
    # ==========================================

    def buscar(self, valor, no=None):

        if no is None:
            no = self.raiz

        i = 0

        while i < len(no.chaves) and valor > no.chaves[i]:
            i += 1

        if i < len(no.chaves) and valor == no.chaves[i]:
            return True

        if no.folha:
            return False

        return self.buscar(valor, no.filhos[i])


    # ==========================================
    # DIVIDE UM NÓ QUANDO ELE FICA CHEIO
    # ==========================================

    def dividir_filho(self, pai, indice):

        t = self.grau_minimo

        filho = pai.filhos[indice]

        novo_no = NoArvoreB(filho.folha)

        chave_meio = filho.chaves[t - 1]

        novo_no.chaves = filho.chaves[t:]
        filho.chaves = filho.chaves[:t - 1]

        if not filho.folha:
            novo_no.filhos = filho.filhos[t:]
            filho.filhos = filho.filhos[:t]

        pai.filhos.insert(indice + 1, novo_no)
        pai.chaves.insert(indice, chave_meio)


    # ==========================================
    # INSERE UM VALOR EM UM NÓ NÃO CHEIO
    # ==========================================

    def inserir_em_no_nao_cheio(self, no, valor):

        i = len(no.chaves) - 1

        # Caso seja folha
        if no.folha:

            no.chaves.append(None)

            while i >= 0 and valor < no.chaves[i]:
                no.chaves[i + 1] = no.chaves[i]
                i -= 1

            no.chaves[i + 1] = valor

        # Caso seja nó interno
        else:

            while i >= 0 and valor < no.chaves[i]:
                i -= 1

            i += 1

            if len(no.filhos[i].chaves) == (2 * self.grau_minimo) - 1:

                self.dividir_filho(no, i)

                if valor > no.chaves[i]:
                    i += 1

            self.inserir_em_no_nao_cheio(no.filhos[i], valor)


    # ==========================================
    # INSERÇÃO PRINCIPAL
    # ==========================================

    def inserir(self, valor):

        raiz = self.raiz

        if len(raiz.chaves) == (2 * self.grau_minimo) - 1:

            nova_raiz = NoArvoreB(False)

            nova_raiz.filhos.append(raiz)

            self.dividir_filho(nova_raiz, 0)

            self.raiz = nova_raiz

            self.inserir_em_no_nao_cheio(nova_raiz, valor)

        else:
            self.inserir_em_no_nao_cheio(raiz, valor)


    # ==========================================
    # PERCORRE A ÁRVORE EM ORDEM
    # ==========================================

    def percorrer_ordenado(self, no=None):

        if no is None:
            no = self.raiz

        resultado = []

        for i in range(len(no.chaves)):

            if not no.folha:
                resultado.extend(
                    self.percorrer_ordenado(no.filhos[i])
                )

            resultado.append(no.chaves[i])

        if not no.folha:
            resultado.extend(
                self.percorrer_ordenado(no.filhos[-1])
            )

        return resultado


# ==========================================
# EXEMPLO DE UTILIZAÇÃO
# ==========================================

arvore = ArvoreB()

valores = [10, 20, 5, 6, 12, 30, 7, 17]

for valor in valores:
    arvore.inserir(valor)

print("Elementos em ordem:")
print(arvore.percorrer_ordenado())

valor_procurado = 12

if arvore.buscar(valor_procurado):
    print(f"\n{valor_procurado} encontrado na árvore.")
else:
    print(f"\n{valor_procurado} não encontrado na árvore.")

````

# **Implementação Prática — Árvore B+** > *Arquiteto de Software: Eliabe Rafael*

Agora vamos colocar a mão na massa! O código abaixo implementa uma **Árvore B+ completa e funcional** com as operações de:
- **Inserção** (com split automático quando um nó enche)
- **Busca** por um valor específico
- **Busca em intervalo** (o grande diferencial da B+)

## **Demonstração Guiada do Código: Árvore B** > *Instrutor de Prática: Reginaldo*

````python
# ============================================================
#   IMPLEMENTAÇÃO DA ÁRVORE B+
#   Ordem M = 4 (cada nó guarda no máximo 3 chaves)
# ============================================================

class No:
    """Representa um nó da Árvore B+ (interno ou folha)."""
    def __init__(self, folha=True):
        self.chaves = []          # valores guardados neste nó
        self.filhos = []          # ponteiros para os filhos (nós internos)
        self.folha = folha        # True = nó folha, False = nó interno
        self.proxima = None       # ponteiro para a próxima folha (lista encadeada)


class ArvoreBMais:
    """Árvore B+ com ordem t (cada nó guarda até 2t-1 chaves)."""

    def __init__(self, t=2):
        self.raiz = No(folha=True)
        self.t = t  # grau mínimo

    # ----------------------------------------------------------
    # BUSCA por um valor específico
    # ----------------------------------------------------------
    def buscar(self, chave, no=None):
        if no is None:
            no = self.raiz

        if no.folha:
            if chave in no.chaves:
                return True
            return False

        # Nó interno: encontra o filho certo para descer
        i = 0
        while i < len(no.chaves) and chave >= no.chaves[i]:
            i += 1
        return self.buscar(chave, no.filhos[i])

    # ----------------------------------------------------------
    # BUSCA EM INTERVALO [inicio, fim]
    # O grande diferencial da B+: percorre as folhas em sequência!
    # ----------------------------------------------------------
    def buscar_intervalo(self, inicio, fim):
        # Passo 1: desce até a folha correta
        no = self.raiz
        while not no.folha:
            i = 0
            while i < len(no.chaves) and inicio >= no.chaves[i]:
                i += 1
            no = no.filhos[i]

        # Passo 2: caminha pelas folhas conectadas coletando resultados
        resultados = []
        while no is not None:
            for chave in no.chaves:
                if inicio <= chave <= fim:
                    resultados.append(chave)
                elif chave > fim:
                    return resultados  # passou do fim, pode parar
            no = no.proxima  # pula direto para a próxima folha!
        return resultados

    # ----------------------------------------------------------
    # INSERÇÃO
    # ----------------------------------------------------------
    def inserir(self, chave):
        raiz = self.raiz
        max_chaves = 2 * self.t - 1

        if len(raiz.chaves) == max_chaves:
            # A raiz está cheia: cria nova raiz e divide
            nova_raiz = No(folha=False)
            nova_raiz.filhos.append(self.raiz)
            self._dividir_filho(nova_raiz, 0)
            self.raiz = nova_raiz

        self._inserir_nao_cheio(self.raiz, chave)

    def _inserir_nao_cheio(self, no, chave):
        max_chaves = 2 * self.t - 1

        if no.folha:
            # Insere mantendo a ordem
            no.chaves.append(chave)
            no.chaves.sort()
        else:
            # Encontra o filho correto
            i = len(no.chaves) - 1
            while i >= 0 and chave < no.chaves[i]:
                i -= 1
            i += 1

            if len(no.filhos[i].chaves) == max_chaves:
                self._dividir_filho(no, i)
                if chave >= no.chaves[i]:
                    i += 1

            self._inserir_nao_cheio(no.filhos[i], chave)

    def _dividir_filho(self, pai, i):
        """Divide o filho i de 'pai' ao meio (Split)."""
        t = self.t
        filho = pai.filhos[i]
        novo = No(folha=filho.folha)

        meio = t - 1
        chave_sobe = filho.chaves[meio]

        # Distribui as chaves
        novo.chaves = filho.chaves[meio:]         # folha: copia o meio
        filho.chaves = filho.chaves[:meio]

        # Distribui os filhos (se for nó interno)
        if not filho.folha:
            novo.filhos = filho.filhos[t:]
            filho.filhos = filho.filhos[:t]
            novo.chaves = filho.chaves[meio + 1:]   # interno: não copia o meio
            filho.chaves = filho.chaves[:meio]
            chave_sobe = filho.chaves[meio] if len(filho.chaves) > meio else chave_sobe

        # Encadeia as folhas
        if filho.folha:
            novo.proxima = filho.proxima
            filho.proxima = novo

        # Sobe a chave do meio para o pai
        pai.chaves.insert(i, chave_sobe)
        pai.filhos.insert(i + 1, novo)

    # ----------------------------------------------------------
    # EXIBIR a estrutura da árvore
    # ----------------------------------------------------------
    def exibir(self, no=None, nivel=0, prefixo='Raiz: '):
        if no is None:
            no = self.raiz
        tipo = '🌿 Folha' if no.folha else '🔀 Interno'
        print(' ' * (nivel * 6) + prefixo + str(no.chaves) + f'  [{tipo}]')
        if not no.folha:
            for idx, filho in enumerate(no.filhos):
                self.exibir(filho, nivel + 1, prefixo=f'Filho {idx}: ')

print('✅ Estrutura da Árvore B+ carregada com sucesso!')

# ==========================================
# EXEMPLO DE UTILIZAÇÃO DA ÁRVORE B+
# ==========================================

# Cria a árvore
arvore = ArvoreBMais()

# Valores que serão inseridos
valores = [10, 20, 5, 15, 25, 30, 35, 40, 45, 50]

# Inserção dos valores
for valor in valores:
    arvore.inserir(valor)

# Exibe a estrutura da árvore
print("\n=== Estrutura da Árvore B+ ===")
arvore.exibir()

# Busca por um valor específico
valor_procurado = 25

print("\n=== Busca por Valor ===")
if arvore.buscar(valor_procurado):
    print(f"Valor {valor_procurado} encontrado.")
else:
    print(f"Valor {valor_procurado} não encontrado.")

# Busca por intervalo
inicio = 15
fim = 40

print("\n=== Busca por Intervalo ===")
resultado = arvore.buscar_intervalo(inicio, fim)

print(f"Valores entre {inicio} e {fim}:")
print(resultado)
````

| Conceito | Explicação Rápida |
|---|---|
| **Árvore B** | Cada nó guarda vários valores. Dados ficam em **todos** os nós. |
| **Árvore B+** | Dados ficam **só nas folhas**. Folhas são conectadas entre si. |
| **Split** | Quando um nó enche, ele se divide. O valor do meio sobe para o pai. |
| **O(log n)** | As duas árvores fazem buscas em tempo logarítmico — muito rápido mesmo com milhões de dados. |
| **Busca em intervalo** | Ponto forte da B+: desce uma vez e caminha pelas folhas em linha reta. |
| **Uso no mercado** | MySQL e PostgreSQL usam B+ para índices. NTFS usa B para o sistema de arquivos. |
